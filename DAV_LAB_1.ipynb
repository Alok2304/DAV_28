### Pandas :

##### The pandas library in Python is a powerful open-source data manipulation and analysis tool that provides easy-to-use data structures and functions designed for working with structured data. It is built on top of the NumPy library and is particularly well-suited for handling labeled and tabular data. The primary data structures in pandas are Series and DataFrame.


### Numpy :

##### NumPy, short for Numerical Python, is a fundamental library for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with an extensive collection of high-level mathematical functions to operate on these arrays. NumPy serves as the foundation for numerous other scientific computing libraries in Python. Its core data structure is the ndarray, an efficient and flexible array that enables operations on large datasets with speed.

### Tensorflow :

##### TensorFlow is a powerful open-source machine learning library developed by the Google Brain team. It is widely used for building and training deep learning models. TensorFlow provides a flexible and comprehensive ecosystem for various machine learning tasks, including neural networks, natural language processing, image recognition, and more.

### PyTorch :

##### PyTorch is a popular open-source machine learning library developed by Facebook's AI Research lab (FAIR). It is widely used for building and training deep learning models, offering a dynamic computational graph that allows for more flexibility compared to static graph libraries. PyTorch's dynamic nature facilitates easier debugging and experimentation.

### BeautifulSoup :

##### Beautiful Soup is a Python library used for web scraping purposes to pull the data out of HTML and XML files. It provides tools for parsing HTML and XML documents, navigating the parse tree, and searching for specific elements. Beautiful Soup creates a parse tree from the page's source code, which can be accessed and manipulated using Python code.

### Dplyr :

##### The `dplyr` package is a popular data manipulation library in R, designed to streamline and simplify the process of working with data frames. Developed by Hadley Wickham, `dplyr` provides a set of functions that make common data manipulation tasks more intuitive and efficient. Key functions in `dplyr` include `filter()` for subsetting rows based on conditions, `select()` for choosing specific columns, `mutate()` for creating or modifying columns, `arrange()` for sorting rows, and `summarize()` for aggregating data.




### Tseries :

##### The `tseries` package in R is a library focused on time series analysis. Developed by Adrian Trapletti and Kurt Hornik, this package provides a range of tools and functions for analyzing and modeling time series data. It includes functions for time series analysis, unit root tests, and various statistical procedures relevant to time-dependent data.

### Ggplot2 :

##### The `ggplot2` package in R, developed by Hadley Wickham, is a powerful and popular data visualization library that follows the principles of the Grammar of Graphics. It allows users to create complex and customized plots with a concise and consistent syntax. Central to `ggplot2` is the concept of layers, where each layer represents a different component of the plot, such as data, aesthetics, and geometry.

### Readr :

##### The `readr` package in R, part of the tidyverse ecosystem developed by Hadley Wickham, is designed for efficient and user-friendly reading of rectangular data files, such as CSV and TSV files. It provides functions that are faster than base R functions for reading data while maintaining a consistent and straightforward interface. The `readr` functions prioritize speed, making them well-suited for handling large datasets.

### StringR :

##### The `stringr` package in R, also part of the tidyverse, is developed by Hadley Wickham and is specifically designed for efficient and consistent string manipulation. It provides a set of functions that simplify common string operations, making it easier for users to work with character data. The functions in `stringr` follow a consistent naming convention and syntax, contributing to a more intuitive and readable code.

```r
# Pandas code to explore basics

import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

# Displaying the DataFrame
print("Original DataFrame:")
print(df)

# Accessing specific columns
print("\nColumn 'Name':")
print(df['Name'])

# Adding a new column
df['Salary'] = [50000, 60000, 45000]

# Displaying the updated DataFrame
print("\nUpdated DataFrame:")
print(df)

# Basic statistics on the numerical columns
print("\nStatistics:")
print(df.describe())
```

```r
# Numpy code to explore basics

import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Displaying the array
print("Original Array:")
print(arr)

# Performing basic operations
mean_value = np.mean(arr)
sum_value = np.sum(arr)
square_root = np.sqrt(arr)
elementwise_square = arr ** 2

# Displaying the results
print("\nMean:", mean_value)
print("Sum:", sum_value)
print("Square Root:", square_root)
print("Elementwise Square:", elementwise_square)
```
